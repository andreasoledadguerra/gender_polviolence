```python
import geopandas as gpd
import pandas as pd

df = pd.read_csv("gender_Sep27-1.csv")
```

```python
df_latlon = df[["latitude", "longitude"]]
```

```python
gdf = gpd.GeoDataFrame(geometry=gpd.points_from_xy(df_latlon["longitude"], df_latlon["latitude"]), crs=4326)
```

```python
gdf.plot()
```
